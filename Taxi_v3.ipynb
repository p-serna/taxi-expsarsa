## We initialize the game and a default agent

```python
env = gym.make('Taxi-v3')
```

```python
from sarsa.monitor import interact
from collections import defaultdict
```

```python
agent = Agent(alpha = 0.15, epsilon = 1e-3, 
              epsilon_decay = 0.9995, alpha_decay = 1.0,
              epsilon_min = 1e-6, alpha_min = 0.3, gamma = 1.0,
             gamma_decay = 1.0, gamma_min = 0.5)
```

```python
avg_rewards, best_avg_reward = interact(env, agent, num_episodes = 20000)
```

```python
agent = Agent(alpha = 0.1, epsilon = 1e-3, 
              epsilon_decay = 0.9995, alpha_decay = 0.999,
              epsilon_min = 1e-6, alpha_min = 0.01, gamma = 1.0,
             gamma_decay = 1.0, gamma_min = 0.5)
```

```python
avg_rewards, best_avg_reward = interact(env, agent, num_episodes = 20000)
```

```python
plt.plot(avg_rewards,'.-')
plt.xlabel('episodes')
plt.ylabel('avg reward')
plt.ylim(0,10)
```
